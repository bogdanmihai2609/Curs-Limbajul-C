Limbajul C este un limbaj de programare imperativ, compilat și de uz general, dezvoltat la Bell Labs de Dennis Ritchie între 1972 și 1973. Este unul dintre cele mai vechi și mai influente limbaje de programare și stă la baza multor alte limbaje moderne.



Principalele caracteristici ale limbajului C includ:


1.   Eficiență și performanță
2.   Limbaj de nivel mediu
3.   Portabilitate
4.   Orientat pe Proceduri
5.   Bază pentru Alte Limbaje

Următorul exemplu ilustrează arhitectura de bază a codului sursă în C, expunând elementele necesare pentru afișarea unui mesaj standard.


```python
int main()
{
    printf("Hello, World!");
    return 0;
}
```

Acest cod reprezintă funcția main() a unui program C:

* **int main()**: Funcția principală, punctul de start al execuției programului.
* **printf("Hello, World!")**: Afișează textul "Hello, World!" pe ecran.
* **return 0**: Indică succesul execuției programului.

**Mediul Integrat de Dezvoltare (IDE)**

Pentru a scrie, compila și rula un program în limbajul C, aveți nevoie de un set de instrumente software.

Practica standard în mediul academic și industrial implică utilizarea unui IDE (**Integrated Development Environment** - Mediu Integrat de Dezvoltare).


**Ce este un IDE?**

Un IDE (Mediu Integrat de Dezvoltare) este un program care oferă toate uneltele necesare pentru a crea software, direct într-o singură fereastră. El este format din trei componente esențiale:

* **Editorul de cod**: Locul unde se scriu instrucțiunile.
* **Compilatorul**: Preia codul scris și îl transformă instantaneu în limbaj mașină (1 și 0), pentru ca procesorul să îl poată executa.
* **Debugger**: Instrumentul dedicat analizei și diagnosticării. Prin intermediul său, utilizatorul poate inspecta secvențial codul, instrucțiune cu instrucțiune.

# **Variabile, Tipuri de Date și Constante**

**Variabila** reprezintă o locație de memorie (o adresă fizică în memoria RAM) identificabilă printr-un nume simbolic (identificator), destinată stocării unei valori specifice. Valoarea stocată în această locație de memorie poate fi modificată pe parcursul execuției programului.

În C, variabilele trebuie să aibă un tip specific, care îi spune programului ce fel de date poate stoca variabila.

* int (32 de biți) - stochează numere întregi, cum ar fi 123 sau -123

* float (32 de biți) - stochează numere cu zecimale, cum ar fi 19.99 sau -19.99

* double (64 de biți) - stochează numere cu zecimale cu precizie dublă (poate memora mai multe zecimale exacte decât float)

* char (8 biți) - stochează un singur caracter, cum ar fi 'a' sau 'B'. Caracterele sunt încadrate de ghilimele simple (apostrofuri)

* short (16 biți) - stochează numere întregi de dimensiuni mai mici, ocupând mai puțină memorie

* long (32 de biți) - stochează numere întregi de dimensiuni mari

* boolean (8 biți) - stochează valori logice (true sau false).


Crearea unei variabile nunite "N" cu tipul int, careia i se atribuie valoarea 15:

```python
int N = 15;
```

Creearea unei variabile cu numele initialaNume de tip char, careia i se atribuie valoarea 'M' și afișează rezultatul pe ecran:

```python
char initialaNume = 'M';
printf("Initiala studentului este: ", initialaNume);
```

Creearea unei variabile cu numele esteAdevarat de tip bool:

```python
bool esteAdevarat = true;
```


**Constanta** este o zonă de memorie a cărei valoare este blocată. Aceasta nu mai poate fi schimbată niciodată pe parcursul rulării.
